```python
import pandas as pd

# Define a function to remove outliers using IQR
def remove_outliers_iqr(df, columns, multiplier=2.0):
    df_no_outliers = df.copy()
    
    for column in columns:
        Q1 = df_no_outliers[column].quantile(0.25)
        Q3 = df_no_outliers[column].quantile(0.75)
        IQR = Q3 - Q1

        lower_bound = Q1 - multiplier * IQR
        upper_bound = Q3 + multiplier * IQR

        df_no_outliers = df_no_outliers[
            (df_no_outliers[column] >= lower_bound) & (df_no_outliers[column] <= upper_bound)
        ]
    
    return df_no_outliers

# Specify columns to check for outliers
numeric_columns = ['Sales', 'Customers', 'CompetitionDistance']

# Remove outliers
merged_train_data = remove_outliers_iqr(merged_train_data, numeric_columns)
```

```python
# Precompute holiday dates and periods
holiday_dates = merged_train_data.loc[merged_train_data['StateHoliday'] != '0', 'Date'].unique()
holiday_periods = {}
for date in holiday_dates:
    holiday_periods[date - pd.DateOffset(days=1)] = 'Before'
    holiday_periods[date] = 'During'
    holiday_periods[date + pd.DateOffset(days=1)] = 'After'

# Create a new column for holiday periods
merged_train_data['holiday_period'] = merged_train_data['Date'].map(holiday_periods).fillna('Neither')

# Filter out rows that are not categorized
holiday_sales = merged_train_data[merged_train_data['holiday_period'] != 'Neither']

# Group by holiday period and calculate average sales
sales_summary = holiday_sales.groupby('holiday_period')['Sales'].mean().reset_index()

# Display the sales summary
print(sales_summary)

# Plotting the sales behavior with custom bar width
plt.figure(figsize=(6, 5))
plt.bar(sales_summary['holiday_period'], sales_summary['Sales'], color=['red', 'orange', 'green'], width=0.3)  # Adjust the width here
plt.xlabel('Holiday Period', fontsize=12)
plt.ylabel('Average Sales', fontsize=12)
plt.title('Average Sales Behavior Before, During, and After Holidays', fontsize=14)
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

<font size = "3">

**INTERPRETATIONS:**

Overall, the data suggests that pharmaceutical sales are highest before the holidays, likely due to consumers preparing in advance, and then decrease during the holidays as some stores might be closed and some stores might have stores opened for only few hours, and further increase after holidays that means the sales is picking up and coming back to normal.

```python
import numpy as np

# Step 1: Create an empty 'HolidayPeriod' column and set default value to NaN
merged_train_data['HolidayPeriod'] = np.nan

# Step 2: Create the mask for 'during' holiday period based on StateHoliday being non-zero (i.e., a holiday)
during_mask = merged_train_data['StateHoliday'] != '0'
merged_train_data.loc[during_mask, 'HolidayPeriod'] = 'during'

# Step 3: Create a mask for 'before' holiday period
# Compare the previous day for each row, checking if the day after is a holiday
merged_train_data['PrevDate'] = merged_train_data['Date'] - pd.Timedelta(days=1)
before_mask = merged_train_data['PrevDate'].isin(merged_train_data.loc[during_mask, 'Date']) & (merged_train_data['StateHoliday'] == '0')
merged_train_data.loc[before_mask, 'HolidayPeriod'] = 'before'

# Step 4: Create a mask for 'after' holiday period
# Compare the next day for each row, checking if the day before is a holiday
merged_train_data['NextDate'] = merged_train_data['Date'] + pd.Timedelta(days=1)
after_mask = merged_train_data['NextDate'].isin(merged_train_data.loc[during_mask, 'Date']) & (merged_train_data['StateHoliday'] == '0')
merged_train_data.loc[after_mask, 'HolidayPeriod'] = 'after'

# Step 5: Drop the temporary columns used for date comparison
merged_train_data.drop(['PrevDate', 'NextDate'], axis=1, inplace=True)
```

# Analyze Sales Performance During Promotions

```python
# Group by Store and calculate average sales and customers during promotions
promo_effectiveness = merged_train_data.groupby('Store').agg({
    'Sales': ['mean', lambda x: x[merged_train_data['Promo'] == 1].mean()],
    'Customers': ['mean', lambda x: x[merged_train_data['Promo'] == 1].mean()]
}).reset_index()

# Rename columns for clarity
promo_effectiveness.columns = ['Store', 'Average_Sales', 'Promo_Sales', 'Average_Customers', 'Promo_Customers']

# Calculate the percentage increase in sales and customers during promotions
promo_effectiveness['Sales_Increase'] = (promo_effectiveness['Promo_Sales'] - promo_effectiveness['Average_Sales']) / promo_effectiveness['Average_Sales'] * 100
promo_effectiveness['Customers_Increase'] = (promo_effectiveness['Promo_Customers'] - promo_effectiveness['Average_Customers']) / promo_effectiveness['Average_Customers'] * 100

# Display top stores with the highest increase in sales and customers
top_stores_sales_increase = promo_effectiveness.sort_values(by='Sales_Increase', ascending=False).head(10)
top_stores_customers_increase = promo_effectiveness.sort_values(by='Customers_Increase', ascending=False).head(10)

```

# Trends of customer behavior during store open and closing times

```python
## for this first let us do feature engineering.

# 1. Create Year, Month, Day, WeekOfYear, and IsWeekend features
merged_train_data['Year'] = merged_train_data['Date'].dt.year
merged_train_data['Month'] = merged_train_data['Date'].dt.month
merged_train_data['Day'] = merged_train_data['Date'].dt.day
merged_train_data['WeekOfYear'] = merged_train_data['Date'].dt.isocalendar().week
merged_train_data['IsWeekend'] = merged_train_data['DayOfWeek'].apply(lambda x: 1 if x in [6, 7] else 0)  # 6 = Saturday, 7 = Sunday

# Proxy: Assuming store opens in the morning (e.g., 9 AM) and closes in the evening (e.g., 8 PM)
# Use the 'Open' column and 'DayOfWeek' as proxy variables

# Create a new feature that indicates if the store is open or closed
# Assuming 'Open' = 1 indicates the store is open, and 0 indicates it's closed
merged_train_data['StoreOpen'] = merged_train_data['Open'].apply(lambda x: 'Open' if x == 1 else 'Closed')

# Alternatively, if you want to create proxy features like 'StoreClose'
# You could assume the store closes after a certain time or use some other criteria
merged_train_data['StoreClose'] = merged_train_data['Open'].apply(lambda x: 'Closing' if x == 0 else 'Open')

# View the updated DataFrame with new features
merged_train_data.head()
```

```python
# Group by StoreOpen to compare open vs closed
store_open_analysis = merged_train_data.groupby('StoreOpen').agg({
    'Sales': 'mean',
    'Customers': 'mean'
}).reset_index()

# Plot average sales and customers for open vs closed stores
fig, ax = plt.subplots(1, 2, figsize=(8, 6))

sns.barplot(x='StoreOpen', y='Sales', data=store_open_analysis, ax=ax[0])
ax[0].set_title('Average Sales during Store Open/Close')
ax[0].set_ylabel('Average Sales')

sns.barplot(x='StoreOpen', y='Customers', data=store_open_analysis, ax=ax[1])
ax[1].set_title('Average Customers during Store Open/Close')
ax[1].set_ylabel('Average Customers')

plt.tight_layout()
plt.show()
```

<font size = "3">

**Key Insights:**

1. Clear Correlation: There is a direct correlation between store opening times and both sales and customer activity. When the store is open, both sales and customer counts are significantly higher.
2. No Activity During Closure: As expected, there are no sales or customer activity when the store is closed.

# Find Stores Open on All Weekdays:

```python
# Find stores open all weekdays (DayOfWeek 1-5) by checking if they are always open (Open = 1) for all days
stores_open_all_weekdays = merged_train_data.groupby('Store').filter(
    lambda x: (x.loc[x['DayOfWeek'].isin([1, 2, 3, 4, 5]), 'Open'] == 1).all()
)

# Extract unique store IDs
unique_stores_open_all_weekdays = stores_open_all_weekdays['Store'].unique()

# Check the number of unique stores
print(f"Stores open on all weekdays: {len(unique_stores_open_all_weekdays)}")

# Print the store IDs for further inspection
print(f"Store IDs: {unique_stores_open_all_weekdays}")
```

```python
# Filter weekend sales data
weekend_sales = merged_train_data[merged_train_data['DayOfWeek'].isin([6, 7])]

# Separate stores into two groups
stores_open_all_weekdays_set = set(unique_stores_open_all_weekdays)
weekend_sales['Group'] = weekend_sales['Store'].apply(
    lambda x: 'OpenAllWeekdays' if x in stores_open_all_weekdays_set else 'NotOpenAllWeekdays'
)

# Calculate average weekend sales for both groups
average_weekend_sales = weekend_sales.groupby('Group')['Sales'].mean()

# Print the results
print(f"Average weekend sales for stores open all weekdays: {average_weekend_sales['OpenAllWeekdays']}")
print(f"Average weekend sales for stores not open all weekdays: {average_weekend_sales['NotOpenAllWeekdays']}")
```

```python
# Data preparation
labels = ['Open All Weekdays', 'Not Open All Weekdays']
average_sales = [average_weekend_sales['OpenAllWeekdays'], average_weekend_sales['NotOpenAllWeekdays']]

# Create a smaller figure
plt.figure(figsize=(6, 4))  # Adjust the figure size here (width, height)

# Bar chart with reduced bar width
plt.bar(labels, average_sales, color=['magenta', 'orange'], width=0.2)
plt.title('Average Weekend Sales by Store Group')
plt.ylabel('Average Sales ($)')
plt.xlabel('Store Group')
plt.tight_layout()  # Adjusts the layout to make it neater
plt.show()

```

<font size = "3">

**Interpretation:**

 The analysis reveals a significant difference in average weekend sales between two groups of stores:

**1. Stores Open All Weekdays:** 
    
These stores have an average weekend sales figure of approximately $7359. This suggests that consistent weekday operations contribute to higher customer familiarity and loyalty, which may translate into better sales on weekends.

**2.  Stores Not Open All Weekdays:**

In contrast, stores that do not operate all weekdays have an average weekend sales figure of about $2,811. This lower sales average could indicate that limited operational days lead to reduced customer engagement or lower brand visibility, resulting in fewer sales during weekends.

Overall, the data suggests that stores with regular hours throughout the week likely cultivate a stronger customer base, which positively affects their weekend performance.

# Check how the assortment type affects sales

```python
# Calculate average sales by assortment type
average_sales_by_assortment = merged_train_data.groupby('Assortment')['Sales'].mean().reset_index()

print(average_sales_by_assortment)
```

```python
# Mapping of assortment types to their corresponding labels
assortment_labels = {'a': 'Basic', 'b': 'Extra', 'c': 'Extended'}

# Create the bar chart
plt.figure(figsize=(6, 5))

# Set the bar width
bar_width = 0.4

# Define colors for each assortment type
colors = plt.cm.viridis(np.linspace(0, 1, len(average_sales_by_assortment)))

# Create the bars
bars = plt.bar(average_sales_by_assortment['Assortment'], 
               average_sales_by_assortment['Sales'], 
               color=colors, 
               width=bar_width)

# Add custom labels (Basic, Extra, Extended) to each bar
for bar, assortment_type in zip(bars, average_sales_by_assortment['Assortment']):
    height = bar.get_height()
    label = assortment_labels[assortment_type]  # Get the label from the dictionary
    plt.text(bar.get_x() + bar.get_width() / 2, height, label, 
             ha='center', va='bottom', fontsize=10)

# Add titles and labels
plt.title('Average Sales by Assortment Type')
plt.ylabel('Average Sales ($)')
plt.xlabel('Assortment Type')
plt.xticks(rotation=45)  # Rotate x labels for better readability
plt.tight_layout()  # Adjust layout to prevent clipping
plt.show()

```

<font size = "3">

**Interpretation:**
    
**Assortment Type ‘a’ Basic:** This type has average sales of approximately.

**Assortment Type ‘b’ Extra:** This type shows the lowest average sales.
    
**Assortment Type ‘c’ Extended:** This type has the highest sales.

From this we can see that assortment type ‘c’  that is the Extended type is the most successful in terms of average sales, while type ‘b’ that is extra type  has the least average sales.

# How does the distance to the next competitor affect sales? What if the store and its competitors all happen to be in city centres, does the distance matter in that case?

```python
# Calculate correlation
correlation = merged_train_data['Sales'].corr(merged_train_data['CompetitionDistance'])
print(f"Correlation between Sales and Competition Distance: {correlation:.2f}")
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=merged_train_data, x='CompetitionDistance', y='Sales', hue='StoreType', palette='Set1')
plt.title('Sales vs Competition Distance by Store Type')
plt.xlabel('Competition Distance')
plt.ylabel('Sales')
plt.show()
```

<font size = "3">

<font size = "3">

**Overall Correlation:**
    
The overall correlation between sales and competition distance is very weak (0.01), indicating that distance to the next competitor has little to no effect on sales in general.
    
**City Centre Scenario:**

If all stores and their competitors are located in city centers, where distances are generally shorter due to higher density, the weak correlation suggests that distance may not be an important factor. In such cases, other factors like foot traffic, store visibility, and local demand might have a more significant influence on sales.

```python
# Filter for stores that had competitors opening after a certain year
new_competitors = merged_train_data[
    (merged_train_data['CompetitionOpenSinceYear'].notna()) &
    (merged_train_data['CompetitionOpenSinceYear'] >= merged_train_data['Year'])
]

# Mark 'before' and 'after' competitor opening based on the date
new_competitors['CompetitorPhase'] = np.where(new_competitors['Year'] >= new_competitors['CompetitionOpenSinceYear'], 'after', 'before')

# Group by 'CompetitorPhase' and calculate average sales
sales_by_phase = new_competitors.groupby('CompetitorPhase')['Sales'].mean().reset_index()

# Plot the sales before and after competitors appear with adjusted bar width
plt.figure(figsize=(6, 4))

# Adjust the width of the bars by setting the 'width' parameter
sns.barplot(x='CompetitorPhase', y='Sales', data=sales_by_phase, palette=['green', 'red'], width=0.3)

plt.title('Sales Before and After Competitor Appears')
plt.ylabel('Average Sales')
plt.xlabel('Competitor Phase')
plt.tight_layout()  # Ensure the layout is adjusted for the labels
plt.show()
```

<font size = "3"> 
      
**Interpretation:**
    
**Sales After Competitor Appears:**

The green bar represents the average sales after a competitor appears or becomes operational. This bar is slightly higher compared to the "before" phase, suggesting that, on average, stores performed marginally better after a new competitor appeared.
    
**Sales Before Competitor Appears:**

The red bar represents the average sales before a competitor was operational or appeared. This phase shows slightly lower sales compared to the "after" phase.
