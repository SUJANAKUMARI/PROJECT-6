```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.feature_selection import SelectKBest, f_regression
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
```

```python
df_train = pd.read_csv('df_train.csv')
df_test= pd.read_csv('df_test.csv')
```

```python
df_train.info()
```

```python
df_test.info()
```

```python
# Convert 'Date' column to datetime
df_train['Date'] = pd.to_datetime(df_train['Date'])
df_test['Date'] = pd.to_datetime(df_test['Date'])

df_train['StateHoliday'] = df_train['StateHoliday'].astype('str')

```

```python
# Encode categorical columns using LabelEncoder
label_cols = ['StateHoliday', 'StoreType', 'Assortment', 'PromoInterval']
le = LabelEncoder()

for col in label_cols:
    df_train[col] = le.fit_transform(df_train[col])
    df_test[col] = le.transform(df_test[col])

# Fill any missing values (if any) with 0 or a suitable strategy
df_train.fillna(0, inplace=True)
df_test.fillna(0, inplace=True)
```

<font size = "5">
Feature importance using correlation matrix

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Correlation matrix for numerical features
corr_matrix = df_train.corr()

# Set up the matplotlib figure
plt.figure(figsize=(14, 10))

# Customize the heatmap for better readability
sns.heatmap(
    corr_matrix, 
    annot=True,              # Show correlation values
    fmt=".2f",               # Limit decimal points for annotations
    cmap="coolwarm",         # Choose a color palette
    linewidths=0.5,          # Add some space between the cells
    annot_kws={"size": 10},  # Adjust annotation font size
    cbar_kws={'shrink': 0.75} # Shrink color bar size
)

# Title for the heatmap
plt.title('Correlation Matrix of Features', size=16)
plt.xticks(rotation=45, ha='right', fontsize=12)  # Rotate x-axis labels
plt.yticks(fontsize=12)                          # Adjust y-axis label size

# Display the heatmap
plt.tight_layout()
plt.show()

# Pick features with strong correlation with 'Sales'
important_features = corr_matrix['Sales'].abs().sort_values(ascending=False).index[:10]
print("Top correlated features with Sales:", important_features)
```

<font size = "5">
    
Feature importance from RandomForest

```python
# Split features and target
X = df_train.drop(['Sales', 'Date', 'Customers'], axis=1)  # Drop columns not relevant for training
y = df_train['Sales']

# Train a RandomForestRegressor to find feature importance
rf = RandomForestRegressor(n_estimators=100, random_state=42)
rf.fit(X, y)

# Get feature importance
feature_importance = pd.Series(rf.feature_importances_, index=X.columns).sort_values(ascending=False)

# Plot the top 10 important features
plt.figure(figsize=(12, 6))
feature_importance[:10].plot(kind='barh')
plt.title("Top 10 Feature Importances")
plt.show()

# Select top features
important_features = feature_importance[:10].index
print("Top 10 important features:", important_features)
```

Based on the important features selected through the correlation and random forest and based on the columns present in the test data set here I am selecting few important features.

```python
# Define features to use for training
features_to_use = [
    'Store',
    'Promo', 
    'DayOfWeek', 
    'StateHoliday', 
    'SchoolHoliday', 
    'StoreType', 
    'Assortment', 
    'CompetitionOpenSinceMonth', 
    'CompetitionOpenSinceYear', 
    'Promo2', 
    'Promo2SinceYear', 
    'Promo2SinceWeek'
]

# Prepare the training data
X_train = df_train[features_to_use]
y_train = df_train['Sales']

# Prepare the testing data
X_test = df_test[features_to_use]
```

# 2.2 & 2.3 # Building models with sklearn pipelines & Choose a loss function

<font size = '5'>
    
Basic Pipeline with Scaling and Random Forest

```python
# Step 1: Split your data
X_train_split, X_val, y_train_split, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Step 2: Define the pipeline steps
pipeline = Pipeline([
    ('scaler', StandardScaler()),  # Step 1: Scaling
    ('model', RandomForestRegressor(n_estimators=100, random_state=42))  # Step 2: RandomForest Model
])

# Step 3: Fit the pipeline on training data
pipeline.fit(X_train_split, y_train_split)

# Step 4: Make predictions
y_val_pred = pipeline.predict(X_val)

# Step 5: Evaluate the model
val_mse = mean_squared_error(y_val, y_val_pred)
val_r2 = r2_score(y_val, y_val_pred)

print(f"Validation MSE: {val_mse:.2f}")
print(f"Validation R²: {val_r2:.2f}")
```

PIPELINE FOR LINEAR REGRESSION MODEL

```python
from sklearn.linear_model import LinearRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Split your data
X_train_split, X_val, y_train_split, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Step 2: Define the pipeline steps
pipeline_lr = Pipeline([
    ('scaler', StandardScaler()),  # Step 1: Scaling
    ('model', LinearRegression())  # Step 2: Linear Regression Model
])

# Step 3: Fit the pipeline on training data
pipeline_lr.fit(X_train_split, y_train_split)

# Step 4: Make predictions
y_val_pred_lr = pipeline_lr.predict(X_val)

# Step 5: Evaluate the model
val_mse_lr = mean_squared_error(y_val, y_val_pred_lr)
val_r2_lr = r2_score(y_val, y_val_pred_lr)

print(f"Linear Regression Validation MSE: {val_mse_lr:.2f}")
print(f"Linear Regression Validation R²: {val_r2_lr:.2f}")
```

PIPELINE FOR GRADIENT BOOSTING ALGORITHM

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error, r2_score

# Step 1: Split your data
X_train_split, X_val, y_train_split, y_val = train_test_split(X_train, y_train, test_size=0.2, random_state=42)

# Step 2: Define the pipeline steps
pipeline_gb = Pipeline([
    ('scaler', StandardScaler()),  # Step 1: Scaling
    ('model', GradientBoostingRegressor(n_estimators=100, random_state=42))  # Step 2: Gradient Boosting Model
])

# Step 3: Fit the pipeline on training data
pipeline_gb.fit(X_train_split, y_train_split)

# Step 4: Make predictions
y_val_pred_gb = pipeline_gb.predict(X_val)

# Step 5: Evaluate the model
val_mse_gb = mean_squared_error(y_val, y_val_pred_gb)
val_r2_gb = r2_score(y_val, y_val_pred_gb)

print(f"Gradient Boosting Validation MSE: {val_mse_gb:.2f}")
print(f"Gradient Boosting Validation R²: {val_r2_gb:.2f}")
```

```python
# Define the parameter grid
param_grid = {
    'n_estimators': [100, 200, 300],             # Number of trees
    'max_depth': [10, 20, 30, None],             # Maximum depth of the trees
    'min_samples_split': [2, 5, 10],             # Minimum number of samples required to split a node
    'min_samples_leaf': [1, 2, 4],               # Minimum number of samples at a leaf node
    'max_features': ['auto', 'sqrt', 'log2']     # Number of features to consider at each split
}

# Initialize the RandomForestRegressor
rf = RandomForestRegressor(random_state=42)

# Initialize GridSearchCV with 1-fold cross-validation
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=2, n_jobs=-1, verbose=2, scoring='neg_mean_squared_error')

# Fit the model on the training data
grid_search.fit(X_train_scaled, y_train)

# Print the best parameters found by GridSearchCV
print("Best parameters found: ", grid_search.best_params_)
```
